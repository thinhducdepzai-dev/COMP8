```python
#10
list1 = list(map(int,input().split()))
n = list1[0]
list1.remove(list1[0])
def so_khac_nhau(lst):
  for i in range(n):
    for j in range(i+1,n):
      if lst[i] == lst[j]:
        return False
  return True
if so_khac_nhau(list1):
  print(sum(list1))
else:
  print('Chuoi khong hop le')
```
