```python
#3
n = int(input("nhap so n: "))
def so_nguyen_to(n):
  if n < 2:
    return str(n) + " khong phai la so nguyen to"
  else:
    for i in range(2, int((n**0.5)+1)):
      if n % i == 0:
        return str(n) + " khong phai la so nguyen to"
    return str(n) + " la so nguyen to"
so_nguyen_to(n)
```

```python
#4
n = int(input("nhap so nguyen n: "))
def kiemtra_hoanhao(x):
  total = 0
  for i in range(1,x):
    if x % i == 0:
      total += i
  if total == n:
    return True
  elif total != n:
    return False
kiemtra_hoanhao(n)
```

```python
#5
lst = list(map(int, input().split()))
k = int(input("nhap so nguyen k: "))
def tim_vi_tri(lst, k):
  for i in range(len(lst)):
    if lst[i] == k:
      return i + 1
  else:
    return -1
tim_vi_tri(lst, k)
```

```python
#6
n = int(input("nhap so n: "))
def giai_thua(n):
  total = 1
  for i in range(1, n+1):
    total *= i
  return total
giai_thua(n)
```

```python
#7
a, b = map(float,input("cap so a va b: ").split())
c = str(input("nhap kieu tinh giua a va b: "))
def maytinh_botui(a,b):
  if c == "-":
    return a - b
  elif c == "+":
    return a + b
  elif c == "*":
    return a * b
  elif c == "/":
    return a / b
x = maytinh_botui(a,b)
print(f"{x:.2f}")
```

```python
#8
a, b = map(int, input("nhap 2 so: ").split())
x = str(bin(a)[2:]).zfill(100)
y = str(bin(b)[2:]).zfill(100)
def khoangcach(x,y):
  count = 0
  for c1, c2 in zip(x, y):   # mỗi lần lấy 2 ký tự
    if c1 != c2:
      count += 1
  return count
khoangcach(x,y)
```

```python
#10
a, b = input("nhap vao 2 tu: ").split()
def is_isomorphic(a: str, b: str) -> bool:
    mapAB = {}
    mapBA = {}

    for x, y in zip(a, b):
        if x in mapAB and mapAB[x] != y:
            return False
        if y in mapBA and mapBA[y] != x:
            return False

        mapAB[x] = y
        mapBA[y] = x

    return True
is_isomorphic(a,b)
```
