```python
#14
a = input("nhap day a: ").split()
b = input("nhap day b: ").split()
c = []
for i in range(len(a)):
  for j in range(len(b)):
    if a[i] == b[j]:
      if a[i] not in c:
        c.append(int(a[i]))
print(c)
```
