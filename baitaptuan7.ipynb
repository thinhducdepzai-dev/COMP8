```python
#1
arr = list(map(int,input().split()))
n = int(input())
left,right = 0,len(arr)-1
def binary_search(arr, left, right, target):
  while left <= right:
      mid = (left + right) // 2
      if arr[mid] == target:
        return mid
      elif arr[mid] < target:
        left = mid + 1
      else:
        right = mid - 1
  return -1
print(binary_search(arr, left, right, n))

```

```python
#2
arr = list(map(int,input().split()))
n = int(input())
left,right = 0,len(arr)-1
new_list = []
def count_occurences(arr,x) -> int:
  for i in range(len(arr)):
    if arr[i] == x:
      new_list.append(i)
  return len(new_list)
print(count_occurences(arr,n))
```

```python
#3
#bubble sort
arr = list(map(int, input().split()))
for i in range(len(arr)):
  for j in range(len(arr)-1):
    if arr[j] > arr[j+1]:
      arr[j],arr[j+1] = arr[j+1],arr[j]
    else:
      pass
new_arr = list(map(str, arr))
new_str = ' '.join(new_arr)
print(new_str)
```

```python
#4
a = list(map(int, input().split()))
b = {}
c = []
d = {}
e = {}
f = []
for i in a:
  b[i] = b.get(i,0) + 1
for v in b.values():
  c.append(v)
for k,v in b.items():
  if v == max(c):
    d[k] = 0
for j in d.keys():
  for i in range(len(a)):
    if a[i] == j:
      e[j] = i
for v in e.values():
  f.append(v)
for k,v in e.items():
  if v == min(f):
    print(k)
```

```python
#4
day_so = list(map(int, input().split()))

dem = {}
for so in day_so:
    dem[so] = dem.get(so, 0) + 1

tan_suat_max = max(dem.values())

ket_qua = None
chua_chon = True

for vi_tri, gia_tri in enumerate(day_so):
    if dem[gia_tri] == tan_suat_max and chua_chon:
        ket_qua = gia_tri
        chua_chon = False

print(ket_qua)
```

```python
nums = eval(input())
def length_of_lis_dp(nums):
    if not nums:
        return 0

    dp = [1] * len(nums)

    for i in range(len(nums)):
        for j in range(i):
            if nums[i] > nums[j]:
                dp[i] = max(dp[i], dp[j] + 1)

    return max(dp)
print(length_of_lis_dp(nums))
```

```python
#8
intervals = eval(input())
queries = eval(input())
a = []
b = []
for q in queries:
  for l,r in intervals:
    if l <= q and q <= r:
      a.append(r-l+1)
  if len(a) == 0:
    b.append(-1)
  else:
    b.append(min(a))
    a.clear()
print(b)


```

```python
#10
nums = eval(input())
n = int(input())
lst = []
a = 0
for i in range(len(nums)):
  lst.append(nums[i]%n)
for i in range(len(lst)):
  for j in range(i,len(lst)):
    if sum(lst[i:j]) < n and sum(lst[i:j]) > a:
      a = sum(lst[i:j])
print(a)
```
