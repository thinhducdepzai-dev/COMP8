```python
#1
class Cylinder:
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height

    def dientich_bemat(self):
        return 2 * 3.14 * self.radius * self.height + 2 * 3.14 * self.radius**2

    def the_tich(self):
        return 3.14 * self.radius**2 * self.height

    def inra(self):
        print(f"Dien tich be mat:{self.dientich_bemat():.2f}")
        print(f"The tich:{self.the_tich():.2f}")
c = Cylinder(3,5)
c.inra()
```

```python
#2
class Date:
  def __init__(self, day, month, year):
    self.day = day
    self.month = month
    self.year = year
  def isleapyear(self):
    if self.year % 400 == 0 or (self.year % 4 == 0 and self.year % 100 != 0):
      return True
    else:
      return False
  def isvalid(self):
    if self.month < 1 or self.month > 12:
      return False
    days_in_month = [31,28,31,30,31,30,31,31,30,31,30,31]
    if self.month == 2 and self.isleapyear():
      max_day = 29
    else:
      max_day = days_in_month[self.month - 1]
    return 1 <= self.day <= max_day
  def next_day(self):
    days_in_month = [31,28,31,30,31,30,31,31,30,31,30,31]
    if self.month == 2 and self.isleapyear():
      max_day = 29
    else:
      max_day = days_in_month[self.month - 1]
    if self.day == max_day and self.month != 12:
      self.day = 1
      self.month += 1
    elif self.day < max_day:
      self.day += 1
    elif self.day == 31 and self.month == 12:
      self.day = 1
      self.month = 1
      self.year += 1
  def __str__(self):
    return f'{self.day:02d}/{self.month:02d}/{self.year:02d}'

s = input().strip()

try:
  d,m,y = map(int,s.split('/'))
  date = Date(d,m,y)
except:
  print('INVALID')
  exit()
if not date.isvalid():
  print('INVALID')
else:
  date.next_day()
  print(date)




```

```python
#3
class Point:
    def __init__(self,x,y):
        self.x = x
        self.y = y
    def position(self):
        if self.x == 0 and self.y != 0:
            return "M nam tren truc tung"
        elif self.x != 0 and self.y == 0:
            return "M nam tren truc hoanh"
        elif self.x == self.y == 0:
            return "M la goc toa do"
        else:
            return "M khong phai goc toa do cung nhu khong phai nam tren truc tung hay hoanh"
    def distance(self,other):
        return f"{((other.x - self.x)**2 + (other.y - self.y)**2 )**0.5:.2f}"
x,y = map(int,input("nhap vao x va y: ").split())
a,b = map(int,input("nhap vao a va b: ").split())
M = Point(x,y)
N = Point(a,b)
print(M.position())
print(M.distance(N))
```

```python
#6
w,h,k = map(float,input('nhap vao 3 so thuc lan luot la width, height, kernel size: ').split())
class Rectangle:
  def __init__(self, width, height):
    self.width = w
    self.height = h
  def area(self):
    return self.width * self.height
  def perimeter(self):
    return (self.width + self.height) * 2
  def scale(self, k):
    self.width *= k
    self.height *= k
M = Rectangle(w,h)
M.scale(k)
print(f"{M.area():.2f}")
print(f"{M.perimeter():.2f}")
```

```python
#7
class Fraction:
  def __init__(self, num, den):
    self.num = num
    self.den = den
  def add(self, other):
    num = self.num * other.den + self.den * other.num
    den = self.den * other.den
    return Fraction(num, den)
  def sub(self, other):
    num = self.num * other.den - self.den * other.num
    den = self.den * other.den
    return Fraction(num, den)
  def mul(self, other):
    num = self.num * other.num
    den = self.den * other.den
    return Fraction(num,den)
  def div(self, other):
    num = self.num * other.den
    den = self.den * other.num
    return Fraction(num,den)
  def __str__(self):
    for i in range(2, min(self.num, self.den)+1):
      while self.num % i == 0 and self.den % i == 0:
        self.num //= i
        self.den //= i
    return f'{self.num}/{self.den}'

a, b, dau, c, d = input('nhap vao so a b dau c d: ').split()

so1 = Fraction(int(a), int(b))
so2 = Fraction(int(c), int(d))

if dau == '+':
    kq = so1.add(so2)
elif dau == '-':
    kq = so1.sub(so2)
elif dau == '*':
    kq = so1.mul(so2)
elif dau == '/':
    kq = so1.div(so2)

print(kq)
```

```python
#9
class Student:
  def __init__(self, name_param):
    self.student_name = name_param
    self.scores = {}

  def add_score(self, subject, score):
    self.scores[subject] = float(score)

  def avg(self, n):
    if n == 0 or not self.scores:
      return "0.00"
    return f'{sum(self.scores.values())/n:.2f}'

  def rank(self, n):
    if n == 0 or not self.scores:
      return "Poor"


    average_score = sum(self.scores.values()) / n

    if average_score >= 8:
      return 'Excellent'
    elif 6.5 <= average_score < 8:
      return 'Good'
    elif 5 <= average_score < 6.5:
      return "Average"
    else:
      return "Poor"

name = input('nhap vao cai ten: ')
y = Student(name)
n = int(input('nhap vao so dong: '))
for i in range(n):
  key,value = input().split()
  y.add_score(key,value)
print(f'{y.student_name} {y.avg(n)} {y.rank(n)}')
```

```python
class Book:
  def __init__(self,title, author, year):
    self.title = title
    self.author = author
    self.year = year
class Library:
  def __init__(self):
    self.books = []
  def add(self,book):
    self.books.append(book)

  def count_by_author(self, name):
    count = 0
    for book in self.books:
      if book.author == name:
        count += 1
    return count
  def find_by_year(self, y):
    count = 0
    for book in self.books:
      if book.year == y:
        count += 1
    return count
n = int(input())
library = Library()
for _ in range(n):
    line = input().strip()

    if line.startswith("ADD"):
        _, data = line.split(" ", 1)
        title, author, year = data.split(";")
        year = int(year)
        book = Book(title, author, year)
        library.add(book)

    elif line.startswith("COUNTYEAR"):
        _, year = line.split()
        print(library.find_by_year(int(year)))

    elif line.startswith("COUNT"):
        _, author = line.split(" ", 1)
        print(library.count_by_author(author))
```

```python
#11
import math

class Point3D:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

    def vector_to(self, other):
        return Vector3D(
            other.x - self.x,
            other.y - self.y,
            other.z - self.z
        )


class Vector3D:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

    def dot(self, other):
        return self.x * other.x + self.y * other.y + self.z * other.z

    def cross(self, other):
        return Vector3D(
            self.y * other.z - self.z * other.y,
            self.z * other.x - self.x * other.z,
            self.x * other.y - self.y * other.x
        )

    def length(self):
        return math.sqrt(self.dot(self))


class Plane:
    def __init__(self, A, B, C):
        AB = A.vector_to(B)
        AC = A.vector_to(C)
        self.normal = AB.cross(AC)

    def angle_with(self, other):
        n1 = self.normal
        n2 = other.normal
        cos_phi = abs(n1.dot(n2)) / (n1.length() * n2.length())
        return math.degrees(math.acos(cos_phi))

A = Point3D(*map(float, input().split()))
B = Point3D(*map(float, input().split()))
C = Point3D(*map(float, input().split()))
D = Point3D(*map(float, input().split()))

plane1 = Plane(A, B, C)
plane2 = Plane(B, C, D)

phi = plane1.angle_with(plane2)

print(f"{phi:.2f}")
```
